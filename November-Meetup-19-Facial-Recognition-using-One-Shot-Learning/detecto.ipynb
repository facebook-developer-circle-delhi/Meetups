```python
import numpy as np
import cv2
# !pip install git+https://github.com/rcmalli/keras-vggface.git
from keras_vggface.vggface import VGGFace
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('trial.jpg')
img2 = np.copy(img)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img2,(x,y),(x+w,y+h),(255,0,0),2)
    andrew = img2[y:y+h, x:x+w]
    
plt.imshow(img2)
```

```python
plt.imshow(andrew)
```

```python
# create a vggface model
model = VGGFace(model='resnet50', include_top=False, input_shape=(224, 224, 3), pooling='avg')
andrew = cv2.resize(andrew, (224,224))
andrew = np.expand_dims(andrew, axis=0)
andrew.shape
```

```python
andF = model.predict(andrew)
andF.shape
```

```python

video_capture = cv2.VideoCapture(1)
font = cv2.FONT_HERSHEY_SIMPLEX
while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        andrew1 = frame[y:y+h, x:x+w]
        andrew1 = cv2.resize(andrew1, (224,224))
        andrew1 = np.expand_dims(andrew1, axis=0)
        andrew1 = model.predict(andrew1)
        d = cosine(andrew1,andF )
        if d < 0.4:
            print("ravi found")
            cv2.putText(frame,'Ravi found',(x,y), font, 1,(255,255,255),2)
        else: 
            cv2.putText(frame,'stranger',(x,y), font, 1,(255,255,255),2)

    # Display the resulting frame
    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```
